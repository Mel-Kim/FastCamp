## Data Split

```python
from sklearn.model_selection import train_test_split
```

```python
scaled_data
```

```python
X=scaled_data.drop('Churn_Yes', axis=1)
y=scaled_data['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(X_test)
```

```python
# List
[1,2,3,4]
```

```python
# Dictionary
{'key1': [1,2,3], 'key2': 'apple'}
```

```python
pd.DataFrame({'actual': y_test, 'pred': pred}).head(30)
```

* 최적의 n의 개수는?

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```

```python
error_list=[]
```

```python
for i in range(1,101):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred=knn.predict(X_test)
    error_list.append(1-accuracy_score(y_test, pred))
```

```python
len(error_list)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(1,101), y=error_list, marker='o', markersize=10, markerfacecolor='red')
```

```python
min(error_list)
```

```python
error_list.index(min(error_list)) # best n = 54(bc index starts 0)
```

```python
np.array(error_list).argmin()
```

```python
knn = KNeighborsClassifier(n_neighbors=54)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```
