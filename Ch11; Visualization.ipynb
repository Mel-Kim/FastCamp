## 1) matplotlib

```python
y = x ** 2
```

```python
plt.figure(figsize=(20,10))
plt.plot(x, y, 'r')
plt.xlabel('x_label')
plt.ylabel('y_label')
#plt.show()
```

+ 객체지향적인 방법: canvas위에 그리기

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.7, 0.7])  # range(0,1)사이의 크기의 캔버스 생성

axes.plot(y,x)
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('Plot Title')
```

```python
# 캔버스 겹치기
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.2, 0.6, 0.3, 0.3])  

axes1.plot(x, y, 'r')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('Plot Title')

axes2.plot(y, x, 'b')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('Plot Title')
```

```python
# 캔버스 나란히 그리기
fig, axes = plt.subplots()

axes.plot(x, y, 'g')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('Plot Title')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y,'r')  # 1st plot
axes[1].plot(y,x,'b')  # 2nd plot

axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Title1')

axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Title2')
```

## 2) seaborn

```python
data = sns.load_dataset('tips')  # data about tips at the restaurant
```

```python
data.head()
```

```python
sns.distplot(data['total_bill'])
```

```python
sns.distplot(data['total_bill'], kde = False, bins = 10)  # delete line, 막대 갯수 설정
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data)  # total_bill과 tip의 dist
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'hex')  # dist 모양 바꾸기
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'reg')  # 선형관계 표시
```

```python
# scatter plot
sns.pairplot(data)  # continuous col에 대해 scatter plot
```

```python
sns.pairplot(data, hue = 'smoker')  # smoker에 따라 다르게 표시
```

```python
sns.rugplot(data['total_bill'])  # 데이터 분포를 바코드 형태로 표시
```

```python
# barplot
sns.barplot(x = 'smoker', y = 'total_bill', data = data)  # 흡연여부에 따라 y평균값 표시
```

```python
sns.barplot(x = 'smoker', y = 'total_bill', data = data, estimator = np.std) # 흡연여부에 따라 y_std 표시
```

```python
sns.countplot(x = 'sex', data = data)  # 성별 count
```

```python
sns.countplot(data['sex'])
```

```python
# boxplot
sns.boxplot(x = 'day', y = 'total_bill', data = data, palette = 'rainbow')  # IQR 표시
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = data, palette = 'Set2')  # color 변경
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = data, palette = 'rainbow', hue = 'smoker' )  # 흡연자 여부에 따라 표시
```

```python
# violinplot
sns.violinplot(x = 'day', y = 'total_bill', data = data, palette = 'rainbow')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = data, palette = 'rainbow', hue = 'smoker')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = data, palette = 'rainbow', hue = 'smoker', spilt = True)
```

```python
data = sns.load_dataset('tips')
```

```python
data['time'].nunique()
```

```python
# hist
g = sns.FacetGrid(data, col = 'time', row = 'smoker')  # categorical value
g = g.map(plt.hist, 'total_bill') 
```

```python
# scatter
g = sns.FacetGrid(data, col = 'time', row = 'smoker')  # categorical value
g = g.map(plt.scatter, 'total_bill', 'tip') 
```

```python
g = sns.FacetGrid(data, col = 'time', row = 'smoker', hue = 'sex')  # categorical value, 성별에 따라 분류
g = g.map(plt.scatter, 'total_bill', 'tip').add_legend()
```
